## Histogram for continous variable

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].mean()).astype(int)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(df['mpg'], bins=20, color='skyblue', edgecolor='black')
plt.title('MPG Distribution')
plt.xlabel('MPG')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.hist(df['displacement'], bins=20, color='green', edgecolor='black')
plt.title('Displacement Distribution')
plt.xlabel('Displacement')
plt.ylabel('Frequency')

plt.subplot(2, 2, 3)
plt.hist(df['weight'], bins=20, color='orange', edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.hist(df['acceleration'], bins=20, color='pink', edgecolor='black')
plt.title('Acceleration Distribution')
plt.xlabel('Acceleration')
plt.ylabel('Frequency')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='mpg', y='horsepower', data=df)
plt.title('Scatter Plot: MPG vs Horsepower')
plt.xlabel('MPG')
plt.ylabel('Horsepower')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = df['horsepower'].astype(float)
y = df['mpg']
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.title('Relationship between Horsepower and MPG')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

sns.countplot(x='cylinders', data=df, palette='pastel')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.title('Count of Cars by Number of Cylinders')
plt.show()
```
