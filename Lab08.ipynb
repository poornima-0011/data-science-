```python
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
data = open('Book.txt', 'r').read()
```

```python
sentences = sent_tokenize(data)
words = word_tokenize(data)
```

```python
#punctuation
words = [word.lower() for word in words if word.isalpha()]
```

```python
# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
processed_text = ' '.join(lemmatized_words)
```

```python
len(processed_text)
```

```python
words = nltk.word_tokenize(processed_text)
word_freq = nltk.FreqDist(words)
top_words = word_freq.most_common(40)
```

```python
top_words
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
```

```python
car_mask = np.array(Image.open('comment.png'))
```

```python
wordcloud = WordCloud(width=800, height=800, background_color='white', mask=car_mask, stopwords=STOPWORDS).generate_from_frequencies(dict(top_words))
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
```
